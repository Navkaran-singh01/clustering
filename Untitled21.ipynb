```python
#Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
# Visualize the reduced dataset
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='True labels')
plt.show()

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

# Plotting the K-Means clusters
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster labels')
plt.show()

# Evaluation Metrics for K-Means
sil_score_kmeans = silhouette_score(X_scaled, y_kmeans)
db_score_kmeans = davies_bouldin_score(X_scaled, y_kmeans)
calinski_score_kmeans = calinski_harabasz_score(X_scaled, y_kmeans)

# Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
y_agg = agg_clustering.fit_predict(X_scaled)
# Plotting the Agglomerative clusters
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_agg, cmap='viridis')
plt.title('Agglomerative Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster labels')
plt.show()

# Evaluation Metrics for Agglomerative Clustering
sil_score_agg = silhouette_score(X_scaled, y_agg)
db_score_agg = davies_bouldin_score(X_scaled, y_agg)
calinski_score_agg = calinski_harabasz_score(X_scaled, y_agg)

# DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X_scaled)

# Plotting the DBSCAN clusters
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_dbscan, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster labels')
plt.show()

# Evaluation Metrics for DBSCAN
sil_score_dbscan = silhouette_score(X_scaled, y_dbscan)
db_score_dbscan = davies_bouldin_score(X_scaled, y_dbscan)
calinski_score_dbscan = calinski_harabasz_score(X_scaled, y_dbscan)
# Create a summary table of results
results = pd.DataFrame({
    'Algorithm': ['K-Means', 'Agglomerative', 'DBSCAN'],
    'Silhouette Score': [sil_score_kmeans, sil_score_agg, sil_score_dbscan],
    'Davies-Bouldin Score': [db_score_kmeans, db_score_agg, db_score_dbscan],
    'Calinski-Harabasz Score': [calinski_score_kmeans, calinski_score_agg, calinski_score_dbscan]
})

# Show results
results
sil_score_kmeans
```
